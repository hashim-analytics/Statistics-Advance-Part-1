```python
# =======================================
# Statistics Advanced Part 1 - Assignment
# =======================================

# Q1: What is a random variable in probability theory?
# A random variable is a numerical outcome of a random phenomenon.
# Example: Rolling a dice, X = {1,2,3,4,5,6}

# Q2: What are the types of random variables?
# Discrete Random Variable: finite or countable outcomes
# Continuous Random Variable: infinite outcomes in a range

# Q3: What is the difference between discrete and continuous distributions?
# Discrete: probability mass function (PMF)
# Continuous: probability density function (PDF)

# Q4: What are probability distribution functions (PDF)?
# Function that assigns probabilities to each possible outcome
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,100)
pdf = norm.pdf(x, 0, 1)
plt.plot(x,pdf)
plt.title("Q4: Probability Distribution Function (PDF) - Normal")
plt.show()

# Q5: How do cumulative distribution functions (CDF) differ from PDF?
# CDF gives probability that variable <= x
cdf = norm.cdf(x,0,1)
plt.plot(x,cdf)
plt.title("Q5: Cumulative Distribution Function (CDF) - Normal")
plt.show()

# Q6: What is a discrete uniform distribution?
# All outcomes are equally likely
from scipy.stats import randint
x_discrete = np.arange(1,7)
pmf = randint.pmf(x_discrete,1,7)
plt.stem(x_discrete, pmf)
plt.title("Q6: Discrete Uniform Distribution PMF")
plt.show()

# Q7: What are the key properties of a Bernoulli distribution?
# Only two outcomes: success (1) and failure (0)
from scipy.stats import bernoulli
bernoulli_pmf = bernoulli.pmf([0,1],0.5)
print("Q7: Bernoulli PMF for p=0.5:", bernoulli_pmf)

# Q8: What is the binomial distribution, and how is it used?
from scipy.stats import binom
binom_rv = binom.rvs(n=10,p=0.5,size=1000)
plt.hist(binom_rv,bins=11,density=True)
plt.title("Q8: Binomial Distribution Histogram")
plt.show()

# Q9: What is the Poisson distribution and where is it applied?
from scipy.stats import poisson
poisson_rv = poisson.rvs(mu=3,size=1000)
plt.hist(poisson_rv,bins=15,density=True)
plt.title("Q9: Poisson Distribution Histogram")
plt.show()

# Q10: What is a continuous uniform distribution?
# Uniform values over an interval
cont_uniform = np.random.uniform(0,1,1000)
plt.hist(cont_uniform,bins=20,density=True)
plt.title("Q10: Continuous Uniform Distribution")
plt.show()

# Q11: What are the characteristics of a normal distribution?
# Symmetric, bell-shaped, mean=median=mode, defined by mean & std

# Q12: What is the standard normal distribution, and why is it important?
# Normal distribution with mean=0, std=1
x = np.linspace(-4,4,100)
plt.plot(x,norm.pdf(x,0,1))
plt.title("Q12: Standard Normal Distribution")
plt.show()

# Q13: What is the Central Limit Theorem (CLT), and why is it critical?
# CLT: Sample means ~ Normal, even if population not normal
samples = [np.mean(np.random.exponential(scale=1,size=30)) for _ in range(1000)]
plt.hist(samples,bins=30)
plt.title("Q13: CLT Sampling Distribution")
plt.show()

# Q14: How does the CLT relate to the normal distribution?
# Shows that sampling distribution of mean ~ Normal

# Q15: What is the application of statistics in hypothesis testing?
# Compare sample vs population using t-tests, Z-tests, p-values

# Q16: How do you calculate a Z-score, and what does it represent?
data = np.random.normal(50,10,100)
z_score = (data - np.mean(data))/np.std(data)
plt.hist(z_score,bins=20)
plt.title("Q16: Z-scores")
plt.show()

# Q17: What are point estimates and interval estimates in statistics?
# Point estimate: single value (mean)
# Interval estimate: range of values (confidence interval)

# Q18: What is the significance of confidence intervals in statistical analysis?
from scipy.stats import t
sample = np.random.normal(50,5,30)
ci = t.interval(0.95,len(sample)-1,loc=np.mean(sample),scale=np.std(sample,ddof=1)/np.sqrt(len(sample)))
print("Q18: 95% Confidence Interval:", ci)

# Q19: What is the relationship between a Z-score and a confidence interval?
# Z-score determines margin of error for confidence intervals

# Q20: How are Z-scores used to compare different distributions?
# Standardize values to a common scale

# Q21: What are the assumptions for applying CLT?
# Independent samples, random sampling, sufficiently large sample size

# Q22: What is the concept of expected value in a probability distribution?
# Weighted average of all possible values
x = np.array([1,2,3,4,5,6])
p = np.ones(6)/6
expected_value = np.sum(x*p)
print("Q22: Expected Value:", expected_value)

# Q23: How does a probability distribution relate to the expected outcome of a random variable?
# Expected value is calculated from the probability distribution
```

```python
# ===============================
# Q1: Write a Python program to generate a random variable and display its value.
# ===============================
import numpy as np
rand_var = np.random.rand()
print("Q1: Random Variable:", rand_var)

# ===============================
# Q2: Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
# ===============================
import matplotlib.pyplot as plt
from scipy.stats import randint
x = np.arange(1, 7)
pmf = randint.pmf(x, 1, 7)
plt.stem(x, pmf)
plt.title("Q2: PMF of Discrete Uniform Distribution (1-6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()

# ===============================
# Q3: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
# ===============================
from scipy.stats import bernoulli
def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)
print("Q3: Bernoulli PDF p=0.5, x=1:", bernoulli_pdf(0.5, 1))

# ===============================
# Q4: Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
# ===============================
from scipy.stats import binom
n, p = 10, 0.5
binom_data = binom.rvs(n, p, size=1000)
plt.hist(binom_data, bins=n+1, density=True, alpha=0.6, color='g')
plt.title("Q4: Binomial Distribution Histogram")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

# ===============================
# Q5: Create a Poisson distribution and visualize it using Python.
# ===============================
from scipy.stats import poisson
poisson_data = poisson.rvs(mu=3, size=1000)
plt.hist(poisson_data, bins=10, density=True, alpha=0.6, color='b')
plt.title("Q5: Poisson Distribution Histogram")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()

# ===============================
# Q6: Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
# ===============================
x = np.arange(1,7)
cdf = randint.cdf(x, 1, 7)
plt.step(x, cdf, where='mid')
plt.title("Q6: CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("CDF")
plt.show()

# ===============================
# Q7: Generate a continuous uniform distribution using NumPy and visualize it.
# ===============================
cont_uniform = np.random.uniform(0, 1, 1000)
plt.hist(cont_uniform, bins=20, density=True, alpha=0.6, color='orange')
plt.title("Q7: Continuous Uniform Distribution")
plt.show()

# ===============================
# Q8: Simulate data from a normal distribution and plot its histogram.
# ===============================
normal_data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='purple')
plt.title("Q8: Normal Distribution Histogram")
plt.show()

# ===============================
# Q9: Write a Python function to calculate Z-scores from a dataset and plot them.
# ===============================
def z_score(data):
    return (data - np.mean(data))/np.std(data)
z_scores = z_score(normal_data)
plt.hist(z_scores, bins=30, color='c', alpha=0.6)
plt.title("Q9: Z-scores Histogram")
plt.show()

# ===============================
# Q10: Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
# ===============================
samples = [np.mean(np.random.exponential(scale=1, size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, color='m', alpha=0.6)
plt.title("Q10: CLT - Sampling Distribution of Mean")
plt.show()

# ===============================
# Q11: Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
# ===============================
means = [np.mean(np.random.normal(5,2,50)) for _ in range(1000)]
plt.hist(means, bins=30, alpha=0.6, color='g')
plt.title("Q11: CLT - Multiple Samples")
plt.show()

# ===============================
# Q12: Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
# ===============================
from scipy.stats import norm
x = np.linspace(-4,4,100)
y = norm.pdf(x, 0, 1)
plt.plot(x,y)
plt.title("Q12: Standard Normal Distribution")
plt.show()

# ===============================
# Q13: Generate random variables and calculate their corresponding probabilities using the binomial distribution.
# ===============================
x = np.arange(0,11)
probs = binom.pmf(x, n=10, p=0.5)
plt.stem(x, probs)
plt.title("Q13: Binomial Distribution Probabilities")
plt.show()

# ===============================
# Q14: Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
# ===============================
data_point = 8
mean_val = np.mean(normal_data)
std_val = np.std(normal_data)
z = (data_point - mean_val)/std_val
print("Q14: Z-score for", data_point, "=", z)

# ===============================
# Q15: Implement hypothesis testing using Z-statistics for a sample dataset.
# ===============================
sample = np.random.normal(50, 5, 30)
pop_mean = 52
z_stat = (np.mean(sample) - pop_mean)/(np.std(sample)/np.sqrt(len(sample)))
print("Q15: Z-statistic =", z_stat)

# ===============================
# Q16: Create a confidence interval for a dataset using Python and interpret the result.
# ===============================
from scipy.stats import t
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
conf_int = t.interval(0.95, len(sample)-1, loc=sample_mean, scale=sample_std/np.sqrt(len(sample)))
print("Q16: 95% Confidence Interval:", conf_int)

# ===============================
# Q17: Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
# ===============================
data_norm = np.random.normal(100, 15, 50)
ci = t.interval(0.95, len(data_norm)-1, loc=np.mean(data_norm), scale=np.std(data_norm, ddof=1)/np.sqrt(len(data_norm)))
print("Q17: 95% CI for Normal Data:", ci)

# ===============================
# Q18: Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
# ===============================
x = np.linspace(-5,5,100)
pdf_vals = norm.pdf(x, 0, 1)
plt.plot(x, pdf_vals)
plt.title("Q18: PDF of Normal Distribution")
plt.show()

# ===============================
# Q19: Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
# ===============================
x = np.arange(0,15)
cdf_vals = poisson.cdf(x, mu=3)
plt.step(x, cdf_vals, where='mid')
plt.title("Q19: CDF of Poisson Distribution")
plt.show()

# ===============================
# Q20: Simulate a random variable using a continuous uniform distribution and calculate its expected value.
# ===============================
rv = np.random.uniform(0,10,1000)
expected_val = np.mean(rv)
print("Q20: Expected value of Continuous Uniform RV:", expected_val)

# ===============================
# Q21: Write a Python program to compare the standard deviations of two datasets and visualize the difference.
# ===============================
data1 = np.random.normal(5,1,100)
data2 = np.random.normal(5,3,100)
plt.boxplot([data1, data2], labels=['Data1','Data2'])
plt.title("Q21: Compare Std Deviations")
plt.show()

# ===============================
# Q22: Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
# ===============================
q1 = np.percentile(data1,25)
q3 = np.percentile(data1,75)
iqr = q3 - q1
range_val = np.max(data1) - np.min(data1)
print("Q22: Range =", range_val, "IQR =", iqr)

# ===============================
# Q23: Implement Z-score normalization on a dataset and visualize its transformation.
# ===============================
normalized_data = (data1 - np.mean(data1))/np.std(data1)
plt.hist(normalized_data, bins=20, alpha=0.6, color='orange')
plt.title("Q23: Z-score Normalization")
plt.show()

# ===============================
# Q24: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
# ===============================
from scipy.stats import kurtosis
from scipy.stats import skew
print("Q24: Skewness =", skew(data1), "Kurtosis =", kurtosis(data1))
```
